```python
def brand(x):
    if x == 'Intel Core i5' or x == 'Intel Core i7' or x == 'Intel Core i3':
        return x
    else:
        if x.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df['Processor'] = df['Processor'].apply(brand)
```
